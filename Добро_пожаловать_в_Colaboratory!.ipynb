```
import pandas as pd
import seaborn as sns


stories_df = pd.read_csv('/content/final_ff_1.csv', sep = ';')


stories_df['inside_flg'] = (stories_df['geo_country'] == 'RU')*1
agg = stories_df[stories_df['like_tap_flg']==1]['inside_flg'].value_counts()
agg2 = stories_df[stories_df['prosmotr'] == 1]['inside_flg'].value_counts()
agg = agg.rename('liked')
agg2 = agg2.rename('viewed')
t1 = pd.merge(agg, agg2, left_on=['inside_flg'], right_on=['inside_flg'])
t1['share'] = t1['liked'] / t1['viewed']
t1 = t1.sort_values(by = 'share', ascending = False)
t1['share'] = round(t1['share'], 3)



t1.head(1)
#if t1.iloc[0]['share'] > t1.iloc[1]['share']:
#  print("inside country", t1.iloc[0]['share'])
#else:
#  print("outside country", t1.iloc[1]['share'])



```
